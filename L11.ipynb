```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, np.pi*4, 51)
y = np.sin(x)
x1= np.linspace(0,np.pi*4,51)
y1= np.cos(x1)
plt.plot(x1,y1)
plt.plot(x, y,color='g')
plt.title('sin cos Plot lab', c='r',fontsize=25) 
plt.legend(['sin', 'cos']) 
plt.ylim(-2, 2)
plt.ylabel('hi there',fontsize=15)
```

```python
with plt.style.context('dark_background'):
    data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
    data['b'] = data['a'] + 10 * np.random.randn(50)
    data['d'] = np.abs(data['d']) * 100
    plt.hist(data['b'])
    plt.title('hihi', size=30, color='b')
    plt.xlim(-20, 100)
    plt.ylim(0, 12)
```
